```python
x=df.drop('LUNG_CANCER',axis=1)
y=df['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 0)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Now that we have prepared data, we will import Keras and its packages.

#ANN

MODEL BUILDING:


Following steps are involved in the model building


Initialising the ANN


Defining by adding layers


Compiling the ANN


Train the ANN

Setup:

```python
import tensorflow as tf
from tensorflow import keras
```

```python


# Sequential groups a linear stack of layers into a Model,Means Stack of the layers in network.

# Layers are the basic building blocks of neural networks in Keras

# Dense  = Every Neuron connected to next layer's neuron:

    # Input shape or imput_dim - defining the input layer with no of neurons or number of columns of the dataset.
    #units - defining the output layer with no. of neurons.
    #actvation - activation function.
    #kernal_initializer - Initializer for the kernel weights matrix or the way in which weights should be provided to an ANN.
    #bias_initializer: Initializer for the bias vector.






# Initialising the ANN

model=keras.Sequential([
    keras.layers.Dense(units = 10, kernel_initializer = 'normal', activation = 'relu', input_dim = 15),
    keras.layers.Dense(8,kernel_initializer = 'normal',activation ='relu'),
    keras.layers.Dense(units = 4, kernel_initializer = 'normal', activation = 'relu')
])
```

```python
# Adding the output layer
model.add(keras.layers.Dense(units = 1, kernel_initializer = 'normal', activation = 'sigmoid'))
```

```python
# compile - Configures the model for training.



#Compiling ANN
model.compile(optimizer = 'adam',#it helps to find the best values for weight and bias so that loss is low and acc. is high.
              loss = 'binary_crossentropy',#loss function to cal to losses.
              metrics = ['accuracy'])#metrics to be evaluated by the model during training and testing, in this the accuracy parameter means Calculates how often predictions equal labels
```

```python
# epoch - An epoch is an iteration over the entire x and y data provided or is the number of complete passes through the training dataset.

# batch_size - batch_size will default to 32,The batch size is a number of samples processed before the model is updated.
#              The size of a batch must be more than or equal to one and less than or equal to the number of samples in the training dataset.


# example:
# Assume you have a dataset with 200 samples (rows of data) and you choose a batch size of 5 and 1,000 epochs.

# This means that the dataset will be divided into 40 batches, each with five samples. The model weights will be updated after each batch of five samples.

# This also means that one epoch will involve 40 batches or 40 updates to the model.

# With 1,000 epochs, the model will be exposed to or pass through the whole dataset 1,000 times. That is a total of 40,000 batches during the entire training process.


# Fitting the ANN model to the Training set.
model.fit(x_train, y_train, batch_size=100, epochs=150)
```

```python
# Returns the loss value & metrics values(accuracy) for the model in test mode.

model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
